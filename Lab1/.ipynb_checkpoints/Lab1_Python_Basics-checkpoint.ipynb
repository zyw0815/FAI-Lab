## PART 1: Jupyter Notebook & Markdown Basics

### 1.1 What is Jupyter Notebook?

**Jupyter Notebook** is an interactive coding environment where you can:
- Write and execute code in "cells"
- Mix text (markdown), code, and output in a single document
- Experiment and iterate quickly

**Two Modes:**
| Mode | How to enter | What you can do |
|------|-------------|-----------------|
| **Command Mode** | Press `Esc` | Navigate cells, add/delete cells, change cell type |
| **Edit Mode** | Press `Enter` | Type code or text inside the selected cell |

**Common Keyboard Shortcuts:**
| Shortcut | Action |
|----------|--------|
| `Shift + Enter` | Run current cell, move to next |
| `Ctrl + Enter` | Run current cell, stay here |
| `A` (Command Mode) | Insert cell **a**bove |
| `B` (Command Mode) | Insert cell **b**elow |
| `DD` (Command Mode) | **D**elete current cell |
| `M` (Command Mode) | Change to **M**arkdown cell |
| `Y` (Command Mode) | Change to code cell (**Y** for pYthon) |

**Example — what it looks like:**
```
[1]  x = 10
     print(x)
     ──────────
     10               ← output appears here

[2]  y = x + 5
     print(y)
     ──────────
     15

[3]  |               ← cursor here, in Edit Mode (blue border)
```
The number in `[ ]` shows execution order.


### 1.2 Code Cells

You execute code in **code cells** using `Shift + Enter`. The output appears directly below the cell.

**Key point:** Variables persist across cells — later cells can use values defined earlier.

**Example:**
```python
# Cell [1]: define a variable
x = 10
print(f"x is {x}")
# Output: x is 10

# Cell [2]: use the variable from Cell [1]
y = x + 5
print(f"y is {y}")
# Output: y is 15

# Cell [3]: modify x — but y is NOT automatically updated!
x = 20
print(f"x is now {x}")  # Output: x is now 20
print(f"y is still {y}")  # Output: y is still 15
```

⚠️ **Trap:** If you run Cell [3] *before* Cell [2], then `y = x + 5` will use `x = 20`, giving `y = 25` instead of 15. Always watch the `[n]` execution numbers!


`````python
# Simple example: assign a variable
x = 10
print(f"x is {x}")
`````

`````python
# The variable x persists in the next cell!
y = x + 5
print(f"y is {y}")
print(f"x is still {x}")
`````

`````python
# Modify x - it doesn't affect y (unless you recalculate)
x = 20
print(f"x is now {x}")
print(f"y is still {y}")
`````

**⚠️ Execution Order Matters!**

Cells run in the order YOU click them, not top-to-bottom automatically.

**Example — correct order:**
```
[1]  x = 5       → x = 5
[2]  y = x + 3   → y = 8   ✅ correct
```

**Example — wrong order (ran cell 2 before cell 1):**
```
[2]  y = x + 3   → NameError: name 'x' is not defined  ❌
[1]  x = 5
```

**Example — stale variable:**
```
[1]  x = 5
[2]  y = x + 3   → y = 8
[3]  x = 100     ← you changed x later
     # y is still 8, not 103!  ⚠️
```

💡 **Tip:** Use **Kernel → Restart & Run All** to run cells in order from top to bottom and reset all variables.

## PART 2: Python Variables and Types

### 2.1 Variables & Assignment

In Python, you don't need to declare variable types. Just assign a value:

```python
x = 5              # x is now an integer (int)
city = "Arad"      # city is a string (str)
is_goal = False    # is_goal is a boolean (bool)
cost = 3.14        # cost is a float
```

**Multiple assignment** — assign several variables in one line:
```python
a, b, c = 1, 2, 3
print(a, b, c)     # → 1 2 3

# Swap two variables (Python trick!):
a, b = b, a
print(a, b)        # → 2 1
```

**Naming conventions:** Use `snake_case` (lowercase with underscores)
```python
total_cost = 100           # ✅ Good
expandedNodesCount = 42    # ❌ Avoid (camelCase is for Java/C++)
x1 = 5                     # ✅ OK for short-lived variables
```

**Example — AI search context:**
```python
initial_city = "Arad"

print(f"Start: {initial_city}")    # → Start: Arad
print(f"Goal: {goal_city}")        # → Goal: Bucharest
print(f"Distance: {path_distance}") # → Distance: 0
```

### 2.2 Data Types

Python has several basic data types:

| Type | Example | `type()` output | Use Case |
|------|---------|-----------------|----------|
| `int` | `42` | `<class 'int'>` | Counts, indices, node IDs |
| `float` | `3.14` | `<class 'float'>` | Distances, probabilities |
| `str` | `"Arad"` | `<class 'str'>` | City names, descriptions |
| `bool` | `True` | `<class 'bool'>` | Flags, goal tests |
| `NoneType` | `None` | `<class 'NoneType'>` | "No value" / unset |

**Checking types with `type()`:**
```python
cost = 5.5
city = "Bucharest"
visited = True
count = 0
nothing = None

print(type(cost))     # → <class 'float'>
print(type(city))     # → <class 'str'>
print(type(visited))  # → <class 'bool'>
print(type(count))    # → <class 'int'>
print(type(nothing))  # → <class 'NoneType'>
```

**⚠️ Common confusion — `int` vs `float`:**
```python
print(10 / 3)    # → 3.3333...  (float, always)
print(10 // 3)   # → 3          (int, floor division)
print(type(4))   # → int
print(type(4.0)) # → float   ← same number, different type!
```


`````python
# Checking types
cost = 5.5
city = "Bucharest"
visited = True
nodes_expanded = 0

print(f"cost = {cost}, type = {type(cost)}")
print(f"city = {city}, type = {type(city)}")
print(f"visited = {visited}, type = {type(visited)}")
print(f"nodes_expanded = {nodes_expanded}, type = {type(nodes_expanded)}")
`````

**Type Conversion:**

You can convert between types using `int()`, `float()`, `str()`, `bool()`:

```python
# str → int
int("42")        # → 42
int("42.9")      # ❌ ValueError! (can't convert decimal string to int)

# str → float
float("3.14")    # → 3.14
float("42")      # → 42.0

# int/float → str
str(100)         # → "100"
str(3.14)        # → "3.14"

# number → bool
bool(0)          # → False   (zero is False)
bool(1)          # → True    (any nonzero is True)
bool(-99)        # → True

# str → bool
bool("")         # → False   (empty string is False)
bool("hello")    # → True    (non-empty string is True)
```

**Practical example:**
```python
user_input = "140"           # input() always returns str
distance = int(user_input)   # convert to int for arithmetic
cost = distance * 0.5        # 140 * 0.5 = 70.0  (float)
print(str(cost) + " units")  # back to str for concatenation → "70.0 units"
```


`````python
# Type conversion examples
str_number = "42"
int_number = int(str_number)  # "42" -> 42
float_number = float("3.14")  # "3.14" -> 3.14
str_bool = str(True)           # True -> "True"

print(f"str_number = {str_number}, type = {type(str_number)}")
print(f"int_number = {int_number}, type = {type(int_number)}")
print(f"float_number = {float_number}, type = {type(float_number)}")
print(f"str_bool = {str_bool}, type = {type(str_bool)}")
`````

### 2.3 String Operations

**f-strings** (formatted strings) are the most readable way to embed values:

```python
cost = 150
city_a, city_b = "Arad", "Bucharest"
print(f"Route from {city_a} to {city_b} costs {cost} units")
# → Route from Arad to Bucharest costs 150 units

# You can even do arithmetic inside {}
print(f"Double cost: {cost * 2}")   # → Double cost: 300
print(f"π ≈ {3.14159:.2f}")         # → π ≈ 3.14  (2 decimal places)
```

**Useful string methods — each with example:**

| Method | Example | Result |
|--------|---------|--------|
| `len(s)` | `len("Arad")` | `4` |
| `s.upper()` | `"arad".upper()` | `"ARAD"` |
| `s.lower()` | `"ARAD".lower()` | `"arad"` |
| `s.strip()` | `"  hello  ".strip()` | `"hello"` |
| `s.replace(a,b)` | `"A->B".replace("->","→")` | `"A→B"` |
| `s.split(sep)` | `"A,B,C".split(",")` | `["A","B","C"]` |
| `sep.join(lst)` | `"→".join(["A","B","C"])` | `"A→B→C"` |
| `s.startswith(p)` | `"Arad".startswith("A")` | `True` |

**Indexing and slicing:**
```python
path = "Arad → Sibiu"
#       0123456789...

path[0]      # → 'A'           (first character)
path[-1]     # → 'u'           (last character)
path[0:4]    # → 'Arad'        (index 0 up to, not including, 4)
path[7:]     # → 'Sibiu'       (from index 7 to end)
path[:4]     # → 'Arad'        (from start to index 4)
path[::2]    # → 'Aa →Sbu'    (every 2nd character)
```


`````python
# String operations
path_description = "Arad -> Sibiu -> Bucharest"

print(f"Original: {path_description}")
print(f"Length: {len(path_description)}")
print(f"Uppercase: {path_description.upper()}")
print(f"Lowercase: {path_description.lower()}")
print()

# Indexing and slicing
print(f"First character: {path_description[0]}")
print(f"First 4 characters: {path_description[0:4]}")
print(f"Last 10 characters: {path_description[-10:]}")
print()

# Split and join
cities = path_description.split(" -> ")
print(f"Cities in path: {cities}")
print(f"Join cities with comma: {', '.join(cities)}")

# f-string with formatting
cost_per_km = 0.5
distance = 140
total_cost = cost_per_km * distance
print(f"\nDistance: {distance} km")
print(f"Total cost: ${total_cost:.2f}")  # 2 decimal places
`````

### 2.4 Basic Operations

**Arithmetic operators — with examples:**

```python
10 + 3   # → 13    (addition)
10 - 3   # → 7     (subtraction)
10 * 3   # → 30    (multiplication)
10 / 3   # → 3.333...  (true division, always float)
10 // 3  # → 3     (floor division, drop remainder)
10 % 3   # → 1     (modulo: remainder after division)
2 ** 10  # → 1024  (exponentiation: 2^10)
```

**Comparison operators — always return `True` or `False`:**

```python
5 == 5   # → True   (equal)
5 != 3   # → True   (not equal)
5 < 10   # → True   (less than)
5 > 10   # → False  (greater than)
5 <= 5   # → True   (less than or equal)
5 >= 6   # → False  (greater than or equal)
```

**Logical operators:**

```python
True and True    # → True
True and False   # → False
True or False    # → True
not True         # → False

# Practical example:
cost = 15
in_bounds = (0 <= cost <= 20)   # → True  (chaining comparisons)
can_expand = cost < 20 and not False   # → True
```

**⚠️ A common mistake — `=` vs `==`:**
```python
x = 5      # assignment: set x to 5
x == 5     # comparison: is x equal to 5? → True
if x = 5:  # ❌ SyntaxError! You must use == in conditions
```


`````python
# Arithmetic
print("=== Arithmetic ===")
print(f"10 + 3 = {10 + 3}")
print(f"10 - 3 = {10 - 3}")
print(f"10 * 3 = {10 * 3}")
print(f"10 / 3 = {10 / 3}")
print(f"10 // 3 = {10 // 3}")
print(f"10 % 3 = {10 % 3}")
print(f"2 ** 10 = {2 ** 10}")
print()

# Comparison
print("=== Comparison ===")
print(f"5 == 5: {5 == 5}")
print(f"5 != 3: {5 != 3}")
print(f"5 < 10: {5 < 10}")
print(f"5 >= 5: {5 >= 5}")
print()

# Logical
print("=== Logical ===")
print(f"True and True: {True and True}")
print(f"True and False: {True and False}")
print(f"True or False: {True or False}")
print(f"not True: {not True}")
`````

`````python
# Practical example: AI search context
start_state = (0, 0)
goal_state = (4, 4)
current_state = (2, 3)
cost_so_far = 15
max_allowed_cost = 20

# Check if goal reached
is_goal = (current_state == goal_state)
print(f"Goal reached? {is_goal}")

# Check if cost is acceptable
cost_acceptable = (cost_so_far <= max_allowed_cost)
print(f"Cost acceptable? {cost_acceptable}")

# Can expand?
can_expand = cost_acceptable and not is_goal
print(f"Can expand? {can_expand}")
`````

## PART 3: Python Data Structures

Data structures let you organize and work with groups of data efficiently. Different structures are optimal for different tasks.

### 3.1 Lists

A **list** is an **ordered, mutable** collection. Perfect for:
- Frontier (nodes waiting to be explored)
- Path (sequence of cities)
- Any ordered data that might change

**Creating lists:**
```python
frontier = []                         # empty list
path = ["Arad", "Sibiu", "Bucharest"] # list with 3 strings
costs = [140, 80, 101]                # list of numbers
mixed = ["Arad", 140, True]           # Python allows mixed types
```

**Step-by-step: building a frontier**
```python
frontier = []
frontier.append("Arad")     # frontier = ["Arad"]
frontier.append("Zerind")   # frontier = ["Arad", "Zerind"]
frontier.append("Timisoara")# frontier = ["Arad", "Zerind", "Timisoara"]

len(frontier)               # → 3
frontier[0]                 # → "Arad"   (first element, index 0)
frontier[-1]                # → "Timisoara" (last element)
frontier[0:2]               # → ["Arad", "Zerind"]  (slicing)

"Arad" in frontier          # → True   (membership check)
"Sibiu" in frontier         # → False
```

**Key list methods:**
```python
lst = ["A", "B", "C"]

lst.pop()       # removes + returns last → "C"  ; lst = ["A","B"]
lst.pop(0)      # removes + returns first → "A" ; lst = ["B"]
lst.insert(0,"Z")  # insert at position 0    ; lst = ["Z","B"]
lst.remove("B")    # remove first occurrence ; lst = ["Z"]
lst.sort()         # sort in-place  (works for numbers/strings)
sorted(lst)        # returns new sorted list (original unchanged)
```

**Stack (LIFO) vs Queue (FIFO):**
```python
# Stack — last in, first out (DFS uses this)
stack = []
stack.append(1); stack.append(2); stack.append(3)  # [1, 2, 3]
stack.pop()    # → 3  (takes from end)

# Queue — first in, first out (BFS uses this)
queue = []
queue.append("A"); queue.append("B")   # ["A", "B"]
queue.pop(0)   # → "A"  (takes from front)
# Note: for large queues, use collections.deque for efficiency
```


`````python
# List operations
frontier = []  # Empty list
print(f"Empty frontier: {frontier}")

# append() - add to end
frontier.append("A")
frontier.append("B")
frontier.append("C")
print(f"After appends: {frontier}")

# Indexing
print(f"First element: {frontier[0]}")
print(f"Last element: {frontier[-1]}")

# Slicing
print(f"First two: {frontier[0:2]}")

# len()
print(f"Length: {len(frontier)}")

# in operator
print(f"Is 'B' in frontier? {'B' in frontier}")
`````

`````python
# More list methods
frontier = ["A", "B", "C"]

# pop() - remove and return last element (stack behavior)
last = frontier.pop()
print(f"Popped {last}, frontier is now {frontier}")

# pop(0) - remove from front (queue behavior)
first = frontier.pop(0)
print(f"Removed from front {first}, frontier is now {frontier}")

# insert() - insert at specific position
frontier.insert(0, "A")  # Insert "A" at position 0
print(f"After insert: {frontier}")

# remove() - remove first occurrence of value
frontier.remove("A")
print(f"After remove: {frontier}")

# extend() - add multiple elements
frontier.extend(["D", "E"])
print(f"After extend: {frontier}")
`````

`````python
# Lists as stacks (LIFO - Last In First Out)
print("=== Stack Behavior ===")
stack = []
stack.append(1)  # Push 1
stack.append(2)  # Push 2
stack.append(3)  # Push 3
print(f"Stack: {stack}")
popped = stack.pop()  # Pop 3
print(f"Popped: {popped}, stack: {stack}")
print()

# Lists as queues (FIFO - First In First Out)
print("=== Queue Behavior ===")
queue = []
queue.append("A")  # Enqueue A
queue.append("B")  # Enqueue B
queue.append("C")  # Enqueue C
print(f"Queue: {queue}")
first = queue.pop(0)  # Dequeue from front
print(f"Dequeued: {first}, queue: {queue}")
print()

# Sorting
print("=== Sorting ===")
numbers = [3, 1, 4, 1, 5, 9]
numbers.sort()
print(f"Sorted: {numbers}")
`````

`````python
# 2D Lists for grids
print("=== 2D Lists ===")
# Represent a 3x3 grid
grid = [
    [0, 1, 0],
    [1, 0, 1],
    [0, 1, 0]
]
print("Grid:")
for row in grid:
    print(row)

print(f"\nElement at [1][2]: {grid[1][2]}")
print(f"First row: {grid[0]}")
print(f"Second column: {[row[1] for row in grid]}")
`````

### 3.2 Tuples

A **tuple** is an **ordered, immutable (unchangeable)** collection. Perfect for:
- States in search (you don't want to accidentally modify them)
- Dictionary keys (lists can't be keys, but tuples can)
- Returning multiple values from a function

**Creating tuples:**
```python
state = (2, 3)          # 2D position
route = ("Arad", "Sibiu", "Bucharest")
single = (42,)          # ← comma required for 1-element tuple!
empty = ()
```

**Accessing elements — same as lists:**
```python
state = (2, 3)
state[0]        # → 2
state[1]        # → 3
state[-1]       # → 3

row, col = state  # unpacking — very common!
# row = 2, col = 3
```

**Immutability — you CANNOT change a tuple:**
```python
state = (2, 3)
state[0] = 5    # ❌ TypeError: 'tuple' object does not support item assignment

# If you need to "change" it, create a new tuple:
new_state = (state[0] + 1, state[1])   # → (3, 3)
```

**Why immutability is useful for search:**
```python
visited = set()
visited.add((0, 0))    # ✅ tuples can be added to sets
visited.add((2, 3))

costs = {}
costs[(0, 0)] = 0      # ✅ tuples can be dictionary keys
costs[(2, 3)] = 5

# Lists CANNOT be used as set elements or dict keys:
visited.add([0, 0])    # ❌ TypeError: unhashable type: 'list'
```

**Functions returning multiple values use tuples:**
```python
def min_max(lst):
    return min(lst), max(lst)   # returns a tuple (min, max)

low, high = min_max([3, 1, 4, 1, 5])
# low = 1, high = 5
```


`````python
# Tuple basics
state = (2, 3)
print(f"State: {state}")
print(f"Type: {type(state)}")
print(f"Length: {len(state)}")

# Indexing
print(f"First element: {state[0]}")
print(f"Second element: {state[1]}")

# Unpacking
row, col = state
print(f"Unpacked: row={row}, col={col}")

# Tuples are immutable - this will error if uncommented:
# state[0] = 5  # TypeError: 'tuple' object does not support item assignment
`````

`````python
# Why tuples matter: as dictionary keys
print("=== Using Tuples as Keys ===")
distances = {
    ("Arad", "Sibiu"): 140,
    ("Sibiu", "Bucharest"): 101,
    ("Arad", "Zerind"): 75
}

print("Distances between cities:")
for route, distance in distances.items():
    city1, city2 = route
    print(f"{city1} -> {city2}: {distance} km")

print(f"\nDistance from Arad to Sibiu: {distances[('Arad', 'Sibiu')]} km")
`````

`````python
# Parent tracking in search
print("=== Using Dict for Parent Tracking ===")
parent = {}  # Track where we came from
parent["A"] = None  # A is the start
parent["B"] = "A"   # B was reached from A
parent["C"] = "A"   # C was also reached from A
parent["D"] = "B"   # D was reached from B

# Reconstruct path from D to A
def reconstruct_path(state, parent_dict):
    path = []
    current = state
    while current is not None:
        path.append(current)
        current = parent_dict.get(current)
    return path[::-1]  # Reverse to get A -> ... -> D

path = reconstruct_path("D", parent)
print(f"Path from start to D: {path}")
`````

### 3.4 Sets

A **set** is an **unordered** collection of **unique** elements. Perfect for:
- Tracking visited nodes (no duplicates, fast lookup)
- Removing duplicates from a list
- Membership testing at O(1) speed

**Creating sets:**
```python
visited = set()               # empty set ← must use set(),  set{} creates a dict!
closed = {(0,0), (1,1), (2,2)}  # set with 3 tuples, non-empty set — {} is fine here
from_list = set([1, 2, 3, 2, 1])  # → {1, 2, 3}  duplicates removed!
```

**Uniqueness — the killer feature:**
```python
path = ["Arad", "Sibiu", "Arad", "Bucharest", "Arad"]
visited = set(path)
print(visited)  # → {'Arad', 'Sibiu', 'Bucharest'}  (no duplicates!)
print(len(visited))  # → 3
```

**Membership check — O(1) speed (vs O(n) for list!):**
```python
visited = {"Arad", "Sibiu", "Bucharest"}

"Arad" in visited       # → True   (instant, doesn't matter how big the set is)
"Zerind" in visited     # → False
"Zerind" not in visited # → True
```

**Modifying sets:**
```python
visited = set()
visited.add("Arad")     # {'Arad'}
visited.add("Sibiu")    # {'Arad', 'Sibiu'}
visited.add("Arad")     # {'Arad', 'Sibiu'}  ← duplicate ignored silently!

visited.remove("Sibiu") # {'Arad'}  raises KeyError if not found
visited.discard("Zerind")  # no error even if not found
```

**Set math operations:**
```python
explored = {"Arad", "Sibiu", "Zerind"}
frontier = {"Zerind", "Timisoara", "Fagaras"}

explored | frontier    # Union:        {'Arad','Sibiu','Zerind','Timisoara','Fagaras'}
explored & frontier    # Intersection: {'Zerind'}  (in both)
explored - frontier    # Difference:   {'Arad','Sibiu'}  (in explored, not frontier)
explored ^ frontier    # Symmetric diff: all except common {'Arad','Sibiu','Timisoara','Fagaras'}
```

**⚠️ Sets are unordered — don't rely on order:**
```python
s = {3, 1, 2}
print(s)    # might print {1, 2, 3} or {3, 1, 2} — order is not guaranteed!
```


`````python
# Set basics
visited = set()
print(f"Empty set: {visited}")

# add()
visited.add("A")
visited.add("B")
visited.add("C")
visited.add("A")  # Duplicate - won't be added
print(f"After adds: {visited}")
print(f"Length: {len(visited)}")

# in operator - very fast!
print(f"\n'A' in visited: {'A' in visited}")
print(f"'Z' in visited: {'Z' in visited}")

# remove()
visited.remove("B")
print(f"\nAfter removing B: {visited}")
`````

`````python
# Sets eliminate duplicates
print("=== Eliminating Duplicates ===")
path = ["A", "B", "A", "C", "A", "D", "C"]
print(f"Path with duplicates: {path}")

unique_cities = set(path)
print(f"Unique cities: {unique_cities}")
print(f"Count: {len(unique_cities)}")
`````

`````python
# Set operations
print("=== Set Operations ===")
set1 = {"A", "B", "C"}
set2 = {"B", "C", "D"}

print(f"Set 1: {set1}")
print(f"Set 2: {set2}")
print()
print(f"Union (all): {set1 | set2}")
print(f"Intersection (common): {set1 & set2}")
print(f"Difference (in set1 but not set2): {set1 - set2}")
`````

### 3.5 Which Data Structure to Choose?

| Structure | Ordered? | Mutable? | Duplicates? | Fast lookup? |
|-----------|----------|----------|-------------|-------------|
| **List** | ✅ Yes | ✅ Yes | ✅ Allows | ❌ O(n) |
| **Tuple** | ✅ Yes | ❌ No | ✅ Allows | ❌ O(n) |
| **Dict** | ❌ No* | ✅ Yes | ❌ Keys unique | ✅ O(1) keys |
| **Set** | ❌ No | ✅ Yes | ❌ Unique only | ✅ O(1) |

*Python 3.7+ dicts maintain insertion order, but that's an implementation detail.

**Quick decision guide:**

```
Q: Do you need to change elements after creation?
 └─ No → use TUPLE  (e.g., state = (2, 3))
 └─ Yes ↓

Q: Do you need key→value mapping?
 └─ Yes → use DICT  (e.g., graph = {"Arad": ["Sibiu", ...], ...})
 └─ No ↓

Q: Do you need uniqueness / fast membership test?
 └─ Yes → use SET   (e.g., visited = set())
 └─ No → use LIST   (e.g., frontier = [], path = [])
```

**Concrete AI search examples:**
```python
# ✅ State — tuple (immutable, can be dict key / set element)
state = (2, 3)
new_state = (3, 3)

# ✅ Graph — dict (key → value mapping)
graph = {"Arad": {"Sibiu": 140, "Zerind": 75}}

# ✅ Frontier — list (ordered, changes as we explore)
frontier = [state]
frontier.append(new_state)
next_node = frontier.pop(0)   # BFS: take from front

# ✅ Visited — set (unique states, fast O(1) membership)
visited = set()
visited.add(state)
if new_state not in visited:
    frontier.append(new_state)

# ✅ Path reconstruction — dict (child → parent)
parent = {state: None}
parent[new_state] = state
```

## PART 4: Brief OOP Preview

You don't need to *write* OOP code in this course, but some textbooks and AI libraries use it. After this section you'll be able to **read and understand** OOP code when you encounter it.

### 4.1 The Core Idea: Blueprints and Objects

Think of a **class** as a **blueprint** and an **object** as a **thing built from that blueprint**.

```
Blueprint (class)          Things built from it (objects / instances)
─────────────────          ──────────────────────────────────────────
Recipe for a cake    →     the actual cake you baked
Car design drawing   →     your red Toyota, your friend's blue Honda
Student record form  →     Alice's record, Bob's record, Carol's record
```

A class defines:
- **Attributes** — the data each object stores (name, age, cost …)
- **Methods** — the actions each object can perform (greet, move, is_goal …)

Once you define the blueprint once, you can create as many objects as you want, each with its own data but sharing the same structure.


### 4.2 Anatomy of a Class

```python
class Dog:                         # 1. class keyword + name (CamelCase)
    def __init__(self, name, age): # 2. constructor — runs when you create a Dog
        self.name = name           # 3. attribute — each Dog remembers its name
        self.age  = age            # 3. attribute — each Dog remembers its age

    def bark(self):                # 4. method — action the Dog can do
        print(f"{self.name} says: Woof!")

    def is_puppy(self):            # 4. another method
        return self.age < 2        # 5. self.age reads THIS dog's age
```

Key terms:

| Term | Meaning | Example |
|------|---------|---------|
| `class` | Blueprint definition | `class Dog:` |
| **constructor** `__init__` | Runs automatically when creating an object | sets up `name`, `age` |
| **attribute** | Data stored inside the object | `self.name`, `self.age` |
| **method** | Function belonging to the class | `bark()`, `is_puppy()` |
| `self` | "This particular object" — always the first parameter | `self.name` means *my* name |
| **object / instance** | A concrete thing built from the blueprint | `rex`, `buddy` below |



`````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def bark(self):
        print(f"{self.name} says: Woof!")

    def is_puppy(self):
        return self.age < 2

# Create two Dog objects from the same blueprint
rex   = Dog("Rex",   5)
buddy = Dog("Buddy", 1)

# Each object has its OWN data
print(rex.name)          # → Rex
print(buddy.name)        # → Buddy
print(rex.age)           # → 5

# Call methods — each object uses its own data
rex.bark()               # → Rex says: Woof!
buddy.bark()             # → Buddy says: Woof!

print(rex.is_puppy())    # → False  (Rex is 5)
print(buddy.is_puppy())  # → True   (Buddy is 1)
`````

### 4.3 OOP in AI Code

Here is an example closer to what you'll see in AI textbooks — a `SearchProblem` class that bundles a problem's data and logic together:

```python
class SearchProblem:
    def __init__(self, initial, goal, grid_size=5):
        self.initial   = initial    # starting state
        self.goal      = goal       # target state
        self.grid_size = grid_size  # size of the grid

    def is_goal(self, state):
        return state == self.goal

    def get_actions(self, state):
        row, col = state
        n = self.grid_size
        return [(row+dr, col+dc)
                for dr, dc in [(-1,0),(1,0),(0,-1),(0,1)]
                if 0 <= row+dr < n and 0 <= col+dc < n]
```

**How to read it when you see it in textbooks:**
- `self.xxx` → just the object's stored data; read it as a variable
- `def method(self, ...)` → a function attached to the object; the first argument `self` is always the object itself (Python passes it automatically)
- `obj.method(arg)` → call `method` on `obj`, equivalent to `method(obj, arg)`



`````python
class SearchProblem:
    def __init__(self, initial, goal, grid_size=5):
        self.initial   = initial
        self.goal      = goal
        self.grid_size = grid_size

    def is_goal(self, state):
        return state == self.goal

    def get_actions(self, state):
        row, col = state
        n = self.grid_size
        return [(row+dr, col+dc)
                for dr, dc in [(-1,0),(1,0),(0,-1),(0,1)]
                if 0 <= row+dr < n and 0 <= col+dc < n]

# Create a problem object
p = SearchProblem(initial=(0, 0), goal=(4, 4))

# Use the object — dot notation: object.attribute or object.method()
print(f"Start : {p.initial}")           # → (0, 0)
print(f"Goal  : {p.goal}")              # → (4, 4)

print(f"At goal?  {p.is_goal((4, 4))}") # → True
print(f"At start? {p.is_goal((0, 0))}") # → False

print(f"Actions from (0,0): {p.get_actions((0, 0))}")  # → [(1,0),(0,1)]
print(f"Actions from (4,4): {p.get_actions((4, 4))}")  # → [(-1,4),(4,3)] (border)
`````
